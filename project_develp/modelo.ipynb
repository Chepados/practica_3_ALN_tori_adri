```python
import pandas as pd
pd.options.display.max_columns= None
```

```python
audio_features = pd.read_csv('data/audio_features.csv')
long_df = pd.read_csv('data/long_df.csv')

audio_features['file_name'] = audio_features['file_name'].astype(str) # megear las columnas a√±adiendo los datos de audio_features por valore de son_id
long_df['song_id'] = long_df['song_id'].astype(str)
```

```python
merged_df = long_df.merge(audio_features, left_on='song_id', right_on='file_name', how='left')
```
