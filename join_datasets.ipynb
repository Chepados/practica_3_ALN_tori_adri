# Unir datasets

```python
import pandas as pd
import numpy as np
```

```python
#Leemos el dataset
songs_df = pd.read_csv('songs_df.csv')
```

```python
#Eliminamos los duplicados
songs_df.drop_duplicates(subset=["artist", "song", "progression"], inplace=True)
```

```python
#Eliminamos las columnas que no sean útiles
songs_df.drop(columns=["section", "url"], inplace=True)
```

```python
#Eliminamos aquellas canciones cuyas progresiones son muy poco frecuentes
songs_df.drop(songs_df[songs_df.progression.isin(songs_df.progression.value_counts()[songs_df.progression.value_counts() < 5].index)].index, inplace=True)
```

```python
#Eliminamos partes como (Live)
songs_df.song = songs_df.song.str.replace(r"\(.*\)", "", regex=True)
```

```python
songs_df.song = songs_df.song.str.lower().str.strip()
songs_df.artist = songs_df.artist.str.lower().str.strip()
```

```python
#Reseteamos el índice
songs_df.reset_index(inplace=True, drop=True)
```

```python
songs_df
```

# Dataset etiquetas sentimientos y genero

```python
#Leemos el dataset
muse_df = pd.read_csv('muse_v3.csv')
muse_df.head()
```

```python
#Eliminamos las columnas que no nos hagan falta
muse_df.drop(columns=["lastfm_url", "number_of_emotion_tags", "mbid"], inplace=True)
```

```python
#Renombramos la columna track por simmplicidad
muse_df.rename(columns={"track":"song"}, inplace=True)
```

```python
muse_df.dropna(subset=["spotify_id"], inplace=True)
```

```python
muse_df.drop_duplicates(subset=["spotify_id"], inplace=True)
```

```python
muse_df.song = muse_df.song.str.lower().str.strip()
muse_df.artist = muse_df.artist.str.lower().str.strip()
```

```python
muse_df
```

```python
muse_df.seeds = muse_df.seeds.apply(lambda x: x.replace("\'","").replace("[","").replace("]","").strip().split(", "))
```

```python
muse_df.seeds
```

```python
seed_dummies = pd.get_dummies(muse_df.seeds.explode()).groupby(level=0).sum()
seed_dummies
```

```python
few_sentiments = seed_dummies.sum()[seed_dummies.sum() <= 30].index
few_sentiments
```

```python
#Eliminamos aquellas etiquetas que aparecen muy poco
seed_dummies.drop(columns=few_sentiments, inplace=True)
```

```python
#Concatenamos los dummies al dataset
muse_df = pd.concat([muse_df, seed_dummies.groupby(seed_dummies.index).sum()], axis=1)
```

```python
#Eliminamos las filas que no tengan ninguna etiqueta
empty_sentiment_rows = seed_dummies[seed_dummies.sum(axis=1) == 0].index
muse_df.drop(empty_sentiment_rows, inplace=True)
```

```python
#Eliminamos la columna seeds

muse_df.drop(columns=["seeds"], inplace=True)
```

```python
#Vemos si hay valores nulos
muse_df.isna().sum().sort_values(ascending=False)
```

```python
#Eliminamos partes como (Live)
muse_df.song = muse_df.song.str.replace(r"\(.*\)", "", regex=True)
```

```python
#Eliminamos canciones de las que no se conozca el nombre
muse_df.drop(muse_df[muse_df.song.str.contains("untitled", case=False)].index, inplace=True)
```

```python
muse_df.reset_index(inplace=True, drop=True)
```

```python
muse_df
```

## Juntamos ambos datasets

```python
merged_df = pd.merge(songs_df, muse_df, on=["song", "artist"], how="inner")
```

```python
tags = merged_df.loc[:, "acerbic":]

cols_to_drop = tags.columns[(tags.sum() == 0)]

cols_to_drop
```

```python
merged_df.drop(columns=cols_to_drop, inplace=True)
```

Este es el dataset que vamos a utilizar para entrenar a nuestra red neuronal
